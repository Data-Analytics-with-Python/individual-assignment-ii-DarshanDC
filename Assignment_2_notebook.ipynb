```python
import pandas as pd
import plotly.express as px

df = data[['NumWebVisitsMonth','NumWebPurchases','Income','Response',
           'Education','Marital_Status']].copy()

df['Income'] = pd.to_numeric(df['Income'], errors='coerce')
df = df.dropna(subset=['NumWebVisitsMonth','NumWebPurchases','Income','Response'])
df['Response'] = df['Response'].map({1:'Accepted', 0:'Not accepted'}).astype('category')

fig = px.scatter(
    df,
    x='NumWebVisitsMonth',
    y='NumWebPurchases',
    size='Income',
    color='Response',
    hover_data={
        'Income': ':.0f',
        'Education': True,
        'Marital_Status': True,
        'NumWebVisitsMonth': True,
        'NumWebPurchases': True
    },
    title='Website Visits vs. Online Purchases — Income & Campaign Response',
    labels={
        'NumWebVisitsMonth': 'Number of Website Visits (Last Month)',
        'NumWebPurchases': 'Number of Online Purchases',
        'Income': 'Customer Income',
        'Response': 'Campaign Response'
    }
)

fig.update_traces(marker=dict(opacity=0.7, line=dict(width=0.5, color='DarkSlateGrey')))
fig.update_traces(marker_sizemin=4)
fig.update_layout(legend_title_text='Campaign Response')
fig.show()
fig = px.scatter(
    df, x='NumWebVisitsMonth', y='NumWebPurchases',
    size='Income', color='Response', trendline='ols',
    hover_data=['Income','Education','Marital_Status']
)
```

## Part B (30 points)

In this part, we will study customers' pattern/distribution of spendings across different types of customers. We will visualize such patterns for different customer groups using **bar charts**.  

**Hint**: In this exercise, you may need to use the `df.melt()` function to convert a wide table into a long table. You can find an example of this function in the assignment repository (see **melt_example.ipynb**).

### Task B.1 (15 points)

i. Consider that customers are grouped by their **education levels**. Develop a bar chart with several sets of bars, in which:

* Each set of bars (i.e.,bar of the same color) represent the spendings by customers in different groups on a spefic product category. For example, spending by customers "Graduation", "Master", "PhD", ..., on *fish products*.  
* The chart should allow you to easily visuallize both the **total spending** by customers in a group on all product categories and the **composition** of their spending on those product categories.

ii. Further, answer the following questions in a **text (markdown) cell** and make sure to explain your answer based on the chart:
* Which customer group contributes the highest revenue?

**Hint**: By default, each row in the dataset is shown as a separate rectangle in the bar, and these rectangles have borders that can make the chart look cluttered. In that case, use `fig.update_traces(opacity=1, marker_line_width=0)` to remove the border lines.

```python
import pandas as pd
import plotly.express as px
spending_cols = [
    'MntWines','MntFruits','MntMeatProducts',
    'MntFishProducts','MntSweetProducts','MntGoldProds'
]

dfb = data[['Education'] + spending_cols].copy()
melted = dfb.melt(
    id_vars='Education',
    value_vars=spending_cols,
    var_name='ProductCategory',
    value_name='Spending'
)
melted['Spending'] = pd.to_numeric(melted['Spending'], errors='coerce').fillna(0)
agg = (melted
       .groupby(['Education','ProductCategory'], as_index=False)['Spending']
       .sum())
fig1 = px.bar(
    agg, x='ProductCategory', y='Spending', color='Education',
    barmode='group',
    title='Total Spending by Product Category — grouped by Education',
    labels={'ProductCategory':'Product Category','Spending':'Total Spending ($)'}
)
fig1.update_traces(opacity=1, marker_line_width=0)
fig1.show()
fig2 = px.bar(
    agg, x='Education', y='Spending', color='ProductCategory',
    barmode='stack',
    title='Spending Composition by Education (stacked by category)',
    labels={'Spending':'Total Spending ($)','Education':'Education'}
)
fig2.update_traces(opacity=1, marker_line_width=0)
fig2.show()
totals = agg.groupby('Education', as_index=False)['Spending'].sum().rename(columns={'Spending':'EduTotal'})
perc = agg.merge(totals, on='Education')
perc['Pct'] = 100*perc['Spending']/perc['EduTotal']

fig2p = px.bar(
    perc, x='Education', y='Pct', color='ProductCategory',
    barmode='stack', text_auto='.0f',
    title='Spending Composition by Education (% of each education group)',
    labels={'Pct':'% of Group Spend'}
)
fig2p.update_traces(marker_line_width=0)
fig2p.show()
top_edu = (agg.groupby('Education', as_index=False)['Spending']
             .sum()
             .sort_values('Spending', ascending=False))
top_edu

```

Education Group Contributes the highest to the Revenue.


### Task B.2 (15 points)

i. Consider that customers are grouped by their **marital statuses**. Develop a bar chart with several sets of bars, in which:

* Each set of bars (bars of the same color) represent the spendings by customers from different groups on a specific product category. For example, customers of "single", "married", ..., on Fruits.
* Within each group, the chart should allow you to easily compare the spending of customers on different product categories (e.g., idenfity the highest spending category by single customers).

ii. Further, answer the following questions in a **text (markdown) cell** and make sure to explain your answer based on the chart:
* For each marital status group, which category of product do they spend the highest amount on?
* Does the marital status affect customers' spending distribution across the product categories?  How?

```python
import pandas as pd
import plotly.express as px

spending_cols = [
    'MntWines','MntFruits','MntMeatProducts',
    'MntFishProducts','MntSweetProducts','MntGoldProds'
]

dfm = data[['Marital_Status'] + spending_cols].copy()

melted_m = dfm.melt(
    id_vars='Marital_Status',
    value_vars=spending_cols,
    var_name='ProductCategory',
    value_name='Spending'
)

agg_m = (
    melted_m.groupby(['Marital_Status','ProductCategory'], as_index=False)['Spending']
    .sum()
)

fig = px.bar(
    agg_m,
    x='ProductCategory',
    y='Spending',
    color='Marital_Status',
    barmode='group',
    title='Total Spending by Product Category — grouped by Marital Status',
    labels={'ProductCategory':'Product Category','Spending':'Total Spending ($)'}
)
fig.update_traces(opacity=1, marker_line_width=0)
fig.show()

top_marital = (
    agg_m.groupby('Marital_Status', as_index=False)['Spending']
    .sum()
    .sort_values('Spending', ascending=False)
)
top_marital
```

1. • Married and Together or common-law customers spend the most overall, with Wines being their top category followed by Meat Products.
• Single customers also spend the most on Wines, but at much lower levels, showing smaller budgets yet similar preferences.
• Divorced customers spend the most on Meat Products, slightly ahead of Wines, suggesting practical, household-focused spending.
• Widow/Widower customers have modest overall spending, concentrating mainly on Wines and Fruits.
• Overall pattern: marital status clearly affects spending distribution — partnered customers Married/Together spend significantly more across nearly all categories, while single or widowed customers spend less and allocate more evenly.
2. Yes, couples and families spend significantly more overall and concentrate on premium categories, while singles spend less and more evenly across categories.

## Part C (30 points)

In this part, let us investigate variability in customers' online purchasing behavior.

i. We will use a visualization chart that shows distribution of customers' online purchases. The chart should:
* Show multiple distributions, each being the distribution of number of online purchases by customers of a certain generation (Gen X vs. Millennials,  ignore other generations)
* Allow users to observe the 1st, 2nd, and 3rd quartiles of the distribution, along with any outliers.

ii. Based on the chart, answer the following question:
* How would you comment on the pattern of online purchases across different generations.

iii. In your chart, further divide each distribution into subgroups of customers based on the **number of kids at home** (not including teens).
How would you comment on the pattern of online purchases across the different subgroups of customers?




```python
import numpy as np
import pandas as pd
import plotly.express as px

df = data[['Year_Birth','NumWebPurchases','Kidhome']].copy()
df['Generation'] = pd.cut(
    df['Year_Birth'],
    bins=[-np.inf, 1964, 1980, np.inf],
    labels=['Other','Gen X','Millennial']
)

df = df[df['Generation'].isin(['Gen X','Millennial'])].copy()

fig = px.box(
    df, x='Generation', y='NumWebPurchases',
    points='all', color='Generation',
    category_orders={'Generation':['Gen X','Millennial']},
    title='Distribution of Online Purchases by Generation (Gen X vs Millennials)',
    labels={'NumWebPurchases':'Number of Online Purchases'}
)
fig.update_traces(marker=dict(opacity=0.6, line=dict(width=0.3, color='DarkSlateGrey')))
fig.show()

fig2 = px.box(
    df, x='Generation', y='NumWebPurchases',
    color='Kidhome', points='all',
    category_orders={'Generation':['Gen X','Millennial']},
    title='Online Purchases by Generation and Kids at Home',
    labels={'NumWebPurchases':'Number of Online Purchases','Kidhome':'Kids at Home'}
)
fig2.update_traces(marker=dict(opacity=0.6, line=dict(width=0.3, color='DarkSlateGrey')))
fig2.show()
```


Task C – Insights

C.i – Generations
Millennials show both a higher median and a wider spread of online-purchase counts compared with Gen X. Their boxplot has a higher center line and many upper outliers, meaning a larger share of heavy online shoppers. Gen X customers cluster tightly around lower values, suggesting more moderate digital-buying habits. Overall, Millennials make more online purchases on average and display greater variability in engagement.

C.iii – Generations with Kids at Home
Within every generation, customers with no kids at home record the highest number of online purchases. Those with one or two kids show lower medians and narrower ranges, reflecting time and budget limits. The gap is widest for Millennials—child-free Millennials are the heaviest online buyers overall. In short, having kids reduces online-shopping frequency across both generations.
